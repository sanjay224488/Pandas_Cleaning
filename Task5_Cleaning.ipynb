```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df['Survived'] = df['Survived'].astype(int)
```

```python
def age_group(age):
    if age < 18:
        return "Child"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

df['Age_Group'] = df['Age'].apply(age_group)
```

```python
df.head()
```

```python
df.to_csv("cleaned_data.csv", index=False)
```

### Data Cleaning Steps

1. Imported pandas and numpy libraries.
2. Loaded Titanic dataset using read_csv().
3. Checked data structure using head() and info().
4. Identified missing values using isnull().sum().
5. Filled missing Age values with median.
6. Filled missing Embarked values with mode.
7. Removed duplicate rows to improve data quality.
8. Converted data types for accurate analysis.
9. Created Age_Group column for better categorization.
10. Saved the cleaned dataset as cleaned_data.csv.
